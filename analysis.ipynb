```python
# Import pandas (used for working with data tables)
import pandas as pd

# URL where the dataset is stored online
url = "https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv"

# Read the CSV file from the URL into a DataFrame (table)
df = pd.read_csv(url)

# Display the first 5 rows of the dataset
df.head()
```

```python
# Show the number of rows and columns in the dataset
# Output format: (rows, columns)
df.shape
```

```python
# Display all column names in the dataset
# This helps you understand what variables you’re working with
df.columns
```

```python
# Show a summary of the dataset
# Includes:
# - column names
# - data types (int, float, etc.)
# - number of non-null (non-missing) values
df.info()
```

```python
# Generate descriptive statistics for numerical columns
# Includes:
# - mean (average)
# - std (standard deviation)
# - min and max values
# - quartiles (25%, 50%, 75%)
df.describe()
```

```python
# Count how many times each value appears in the Outcome column
# This helps us understand what we are trying to predict
df['Outcome'].value_counts()
```

```python
# Import visualization library
import matplotlib.pyplot as plt

# Create a bar chart of the target variable
df['Outcome'].value_counts().plot(kind='bar')

# Add title for clarity
plt.title("Distribution of Diabetes Outcome")

# Show the plot
plt.show()
```

```python
# Compute correlation between all variables
correlation = df.corr()

# Display correlation values
correlation
```
